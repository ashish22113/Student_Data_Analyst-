# filling a single coulmn from NaN to "no data"

```python
df["Score"].fillna("no data")
```

```python
filled_data = {"name":
               df["Name"].mode(),
               "Location" : df["Location"].mode()[1],
               "Age" : df["Age"].mean(),
               "Score" : df["Score"].mean(),
               "BloodGroup": df["BloodGroup"].mode(),
               "eligible" :df["eligible"].mean()}
```

```python
df.fillna(filled_data, inplace=True)
```
